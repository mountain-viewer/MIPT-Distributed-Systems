### Графики зависимости  $T(N)$, $S(N)$, $E(N)$ при фиксированном значении $P$

Выберем число потоков $P = 4$, далее для разного числа частиц $N$ запустим наш алгоритм. На основе полученных данных построим графики зависимости $T(N)$, $S(N)$, $E(N)$.

```python
N = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]

T = [0.0005, 0.0012, 0.0031, 0.0092, 0.0150, 0.0726, 0.1462, 0.6958, 1.3592, 6.5503, 12.8552]
S = [2336.4, 1309.3, 2764.5, 2905.1, 2489.7, 2546.9, 2582.1, 2529.8, 2494.4, 2498.2, 2496.1]
E = [0.2, 0.60, 0.36, 0.48, 0.55, 0.52, 0.49, 0.50, 0.50, 0.50, 0.50]
```

Строим графики:

```python
plt.figure(figsize=(10, 6))
plt.plot(N, T)
plt.title("Зависимость T(N) (общее времени работы)")
plt.xlabel("N")
plt.ylabel("T")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(N, S)
plt.title("Зависимость S(N) (среднее число шагов до посещения поглощающего состояния)")
plt.xlabel("N")
plt.ylabel("S")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(N, E)
plt.title("Зависимость E(N) (вероятность посещения правого поглощающего состояния)")
plt.xlabel("N")
plt.ylabel("E")
plt.show()
```

### Графики зависимости  $T(P)$, $S(P)$, $E(P)$ при фиксированном значении $N$

Выберем число частиц $N = 100000$, далее для разного числа потоков $P \in \{1, 2, 4, 8, 16\}$ запустим наш алгоритм. На основе полученных данных построим графики зависимости $T(P)$, $S(P)$, $E(P)$.

```python
P = [1, 2, 4, 8, 16]

T = [3.8502, 1.9086, 1.2832, 1.3107, 1.3365]
S = [2494.4, 2501.0, 2501.6, 2505.2, 2553.0]
E = [0.50, 0.50, 0.50, 0.50, 0.50]
```

Строим графики:

```python
plt.figure(figsize=(10, 6))
plt.plot(P, T)
plt.title("Зависимость T(P) (общее времени работы)")
plt.xlabel("P")
plt.ylabel("T")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(P, S)
plt.title("Зависимость S(P) (среднее число шагов до посещения поглощающего состояния)")
plt.xlabel("P")
plt.ylabel("S")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(P, E)
plt.title("Зависимость E(P) (вероятность посещения правого поглощающего состояния)")
plt.xlabel("P")
plt.ylabel("E")
plt.show()
```

** Вывод: **
По графику $T(P)$ видим, что наибольшая эффективность достигается при четырёх работающих потоках.
