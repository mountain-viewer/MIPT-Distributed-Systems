### Графики зависимости  $T(N)$, $S(N)$, $E(N)$ при фиксированном значении $P$

Выберем число потоков $P = 4$, далее для разного числа частиц $N$ запустим наш алгоритм. На основе полученных данных построим графики зависимости $T(N)$, $S(N)$, $E(N)$.

```python
N = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]

T = [0.000536, 0.001493, 0.003168, 0.017271, 0.025697, 0.109630, 0.279250, 1.412631, 2.282856, 11.019686, 21.085018]
S = [1831, 1820.840, 1799.320, 2531.276, 2248.348, 1993.534, 2386.890, 2330.267, 2035.896, 2013.969, 1992.899]
E = [0.2, 0.560, 0.550, 0.570, 0.498, 0.520, 0.516, 0.513, 0.514, 0.514, 0.516]
```

Строим графики:

```python
plt.figure(figsize=(10, 6))
plt.plot(N, T)
plt.title("Зависимость T(N) (общее времени работы)")
plt.xlabel("N")
plt.ylabel("T")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(N, S)
plt.title("Зависимость S(N) (среднее число шагов до посещения поглощающего состояния)")
plt.xlabel("N")
plt.ylabel("S")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(N, E)
plt.title("Зависимость E(N) (вероятность посещения правого поглощающего состояния)")
plt.xlabel("N")
plt.ylabel("E")
plt.show()
```
