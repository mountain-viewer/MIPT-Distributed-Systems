### Графики зависимости  $T(P)$, $S(P)$, $E(P)$ при фиксированном значении $N$ и $M$

Выберем число элементов массива $N = 100000000$, чтобы провести эксперименты по сравнению скорости работы реализованного алгоритма параллельной сортировки слиянием и скорости работы встроенного в стандартную библиотеку алгоритма быстрой сортировки.

В качестве оптимального размера чанка выберем  $M = 10000$.

Далее, для разного числа потоков $P \in \{1, 2, 4, 8, 16\}$ запустим упомянутые алгоритмы и сравним их времена работы на случайных массивах размера $N$.

```python
P = [1, 2, 4, 8, 16]

quicksort_time = 24.24185
T = [25.59730, 13.54671, 12.91169, 13.11147, 15.35376]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(P, T, color='red', label='parallel merge sort')
plt.plot(P, [quicksort_time] * 5, color='blue', label='quick sort')
plt.legend()
plt.title("Зависимость T(P) (общее времени работы)")
plt.xlabel("P")
plt.ylabel("T")
plt.show()
```

Видим, что наибольшая скорость выполнения параллельного алгоритма достигается при $P = 4$ работающих потоках.

Теперь посчитаем значения ускорения $S(P)$ и эффективности $E(P)$ по полученным данным и построим соответствующие графики.

```python
import numpy as np
```

```python
T = np.array(T)
S = T[0] / T
E = S / P
```

```python
plt.figure(figsize=(10, 6))
plt.plot(P, S)
plt.title("Зависимость S(P) (ускорение)")
plt.xlabel("P")
plt.ylabel("S")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(P, E)
plt.title("Зависимость E(P) (эффективность)")
plt.xlabel("P")
plt.ylabel("E")
plt.show()
```

** Вывод: **

Таким образом, на экспериментальных данных мы получили, что параллельный алгоритм сортировки слияния работает быстрее и эффективнее быстрой сортировки, встроенной в стандартную библиотеку языка `C`.
